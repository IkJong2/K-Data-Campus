## 추천시스템

과거 데이터 기반으로 미래를 예측하는 것 => 기계학습!

어느 순간 추천 = 기계학습이 되버림...

파레토 법칙 (선택과 집중) ... 20% 집중 ..소품종 대량생산

롱테일 법칙 (대중의 힘) ... 80% ... 다품종 소량생산

필터 버블
- 필터링 기법이 너무 늘어나서 그것들이 편향을 만들어냄

추천시스템
- 사용자가 선호할 만한 아이템을 추측함으로써 여러가지 항목들을 ...


- 협업 필터링(User)
    - 기존 사용자(또는 항목) 행동 정보를 분석하여 해당 사용자와 비슷한 성향의 사용자들이 기존에 좋아했던 항목을 추천하는 기술
    - 핵심!! **사용자끼리 유사도를 찾는 것!!**
  
-  
    - 문제점
        - 콜드 스타트(Cold-Start) 문제
            - 처음 데이터가 작을 때, 성능의 문제가 생김...
            - 그래서 실제에서는 데이터가 쌓일 때까지 기다린다.
        - 롱 테일(Long tail) 문제
            - 추천을 위한 충분한 정보를 제공 X... 비대칭적 쏠림 형상
        - 계산량이 많은 알고리즘이면 효율적으로 추천X
        - 독특한 취향의 고객(gray sheep / black sheep) 문제
        
    
- 콘텐츠 기반 필터링(Item)
    - 음악추천과 비슷! ... 안돼! 잘 안되고 힘들다...
    - 아이템기반!
    - **콘텐츠의 내용을 분석해야 하므로 아이템(콘텐츠) 분석 알고리즘이 핵심**
    
    
- 하이브리드(collaborate) 필터링
    - 협업(User) + 콘텐츠(Item)
    
    
- 콜드 스타트(Cold Start) 문제
    - '새로 시작할 때 곤란함'을 위미
    - 아이템 콜드 스타트
    - 이용자 콜드 스타트
    - 시스템 콜드 스타트

### **기술면접 포트폴리오로 정리해야할 것!!**

- 모델 기반 협업 필터링
    - 결국엔 하이브리드 방법을 쓴다!!
    - 계속 정보가 많아지면 많아질 수록 발달할 것이다.
    - 머신러닝을 사용할 것이다! 제일 많이 쓰는 것이 KNN but 숨어 있는 의미 찾기가 힘듬...

오늘 할 것?
- memory base

기본 준비 단계
- Dimension(Column, Feature), Vector(index, row) 들이 마련되어 있어야 한다!
- Similarity Calculation 유사도 계산!!
    - cosine similarity
    - correlation
- 유사도 matrix를 통해 K명을 찾아서 소팅
- K명의 구매목록을 가져와서 내가 산것들 날리기
- 여기서도 전략!
    - 가장 가까운사람의 물품을 가져올 수 있고
    - K명을 싹 다 고려한 물품을 가져올 수도 있다.
    
    
- 아마존 (Item base collaborative system)
    - 그 사람이 샀던 것과 비슷한 물품을 소개

- Contents base collaborative system
    - 컨텐츠 고유의 속성만으로 유사도를 계산하여 테이블을 만든다.
    - 휴대폰,노트북 검색하면 케이스 나오게끔!! 
    
- 알고리즘도 별로 어렵지 않고, 이론도 쉽다... 하지만 정확도를 어떻게 할텨??
    - 매우 Critical한 요소이다.
    - 넷플릭스의 경우 추천받은 사람에게 만족도로 검증했다.
    - 추천시스템 ... 회사에서 엄청 좋아한다.

```python
item['movie_id'] = item['movie_id'].astype('object')
```

```python
def recommendation_user(user_id, k, n=10):
    ##k가 자기랑 비슷한 사람 몇명 찾을지, n이 몇 개까지 영화를 보여 줄 건지
    s = reco_corr.loc[user_id].sort_values(ascending=False).iloc[1:k+1].index
    temp = pd.DataFrame(columns=['user_id','item_id','rating'])
    ##나하고 비슷한 놈 찾는게 s!!
    for j, i in enumerate(s,1):
        t = set(data[data.user_id==i].item_id) - set(data[data.user_id==user_id].item_id)
        
        k = data[(data.item_id.isin(t)) & (data.rating==5) & (data.user_id == i)]
        k.rating = k.rating * (1-0.05*j)
        
        temp = pd.concat([temp,k])
        ##나는 안봤지만 비슷한 영화 찾기!
    
    else:
        temp = temp.rename({"item_id":"movie_id"},axis=1) 
        return list(temp.merge(item[['movie_id', 'movie_title']]).drop_duplicates().groupby('movie_title').mean().sort_values('rating',ascending=False)[:n+1].index)
##판다스에서는 duplicate를 지원
## 복사가 있는지 없는지, 중복이 있는지 없는지를 보여준다!
```
