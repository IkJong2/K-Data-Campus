```python
dust['날짜'] = pd.to_datetime(dust['날짜'])
dust.head()
##datetime이 되는 순간, resampling이 가능함!
##but!!! 시리즈를 넣으면 시리즈이기에, DatetimeIndex로 바꿔줘야한다!!!!
##아직까지 Range Index!! 그래서 resample이 안된다~~
```

```python
dust.info()
```

```python
dust.set_index("날짜", inplace=True)
##이때부터 Resample이 가능~~
```

시간 또한 자동으로 만들어주는 아이 pd.date_range

```python
index = pd.date_range('1/1/2000', periods=9,freq='T')
series = pd.Series(range(9), index=index)
series
##시리즈의 인덱스가 DatetimeIndex이면 resampling이 가능!
```

Resample의 경우 Groupby와 느낌이 비슷

resample(단위).집계함수()

```python
series.resample('3T').sum()
##쌉가능
```

```python
dust.resample('M').mean()
## 이런식으로 평균 값을 낼 수 있다.
## 미세먼지는 예측이 되지 않는다!!
```
