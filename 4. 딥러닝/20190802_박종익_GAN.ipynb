### GAN
- Discriminator
  - real image를 보고 진짜가 무엇인지를 배운다.
  - 동시에 Fake image도 본다.
  - loss가 real(1) fake(0)을 학습한다.
  - Generator를 학습시킬 때에는 기준점! 이 아이는 Update되면 안된다.
  - Epoch을 돌 때에는 다시 Update 되어야함!
  
- Generator
  - random 값을 가지고 fake image를 만드는 아이!
  - fake image가 Discriminator를 통해 0(fake)에 최대한 가깝도록 학습하도록 해야함

- 1step
  - loss 를 D(x) = 1 로 학습, D(G(z)) = 0으로 학습
- 2setp
  - loss를 D(G(z)) = 1 로 학습
  
- Tip
  - Convolution에서 stride를 2로 하면 반으로 줄어듬
  - Deconvolution에서 stride를 2로 하면 두배로 늘어남
  
  **결국 오늘은 Discriminator 와 Generator에서 학습하고**
  
 **Genreator에서 fake_image를 속혀서 1로 학습하게 하는것!! **
 

 ### DCGAN
 
 
 - conv와 deconv 를 해가면서 train에서 과정은 같지만 layer를 convolution으로 쌓은 것!

    - domain은 다르다! e.g. 말 얼룩말, 불어 영어

### CycleGAN

  - latent space에 인코딩 디코딩 해가는 것
  - e.g. 여름 겨울, 말 얼룩말 등등
